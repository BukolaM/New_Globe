```python
#importing the required librairies
import pandas as pd
```

```python
#loading and reading the `lesson` dataset
df_L = pd.read_excel('Lesson completion.xlsx')
df_L.head()
```

```python
#loading and reading the `school information` dataset
df_S = pd.read_excel('School_information.xlsx')
df_S.head()
```

```python
#loading and reading the `Pupil attendance` dataset
df_Pa = pd.read_excel('Pupil attendance.xlsx')
df_Pa.head()
```

```python
#loading and reading the `Pupil scores` dataset
df_Ps = pd.read_excel('Pupil scores.xlsx')
df_Ps.head()
```

## Data Cleaning

- #### Merging all the datasets on the unique id(school_id)

```python
df_1 = pd.merge(df_L, df_S, on="school_id", how="inner")
```

```python
df_2 = pd.merge(df_1,  df_Pa, on="school_id", how="inner")
```

```python
df_all = pd.merge(df_2,  df_Ps, on="school_id", how="inner")
df_all.head(50)
```

- The analysis shows that each school and pupil has a unique identification code. However, the subjects taken are the same across all schools, with varying scores.

```python
#checking if there are null values
df_all.isna().sum()
```

## Calculating KPIs

```python
#identifying the unique present records
df_all['present_records'].unique()
```

```python
df_all['attendance_records'].unique()
```

## 1. What is the network-level average Percent Pupils Present 

```python
df_all['percent_pupils_present'] = (df_all['present_records']/df_all['attendance_records']) * 100
df_all.head(5)
```

```python
total_average = round(df_all['percent_pupils_present'].mean(),2)
total_average
```

## 2. What is the average Percent Pupils Present now (use two decimal points)?

```python
#grouping the school id by the average percent_pupils_present 
df_all_average = df_all.groupby(['school_id'])['percent_pupils_present'].mean().reset_index(
            name='percent_pupils_present').sort_values(['school_id'])
df_all_average 
```

```python
school_average = round(df_all_average['percent_pupils_present'].mean(),2)
school_average
```
